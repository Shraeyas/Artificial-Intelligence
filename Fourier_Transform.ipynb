```python
%matplotlib inline
import numpy as np

from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

import math
```

```python
plt.style.use ( 'seaborn-pastel' )
fig = plt.figure ()
ax = plt.axes (xlim = ( -5, 5 ), ylim = ( -3, 3 ) )
line, = ax.plot ( [], [], lw = 3 )

r = 4

def init () :
    line.set_data ( [], [] )
    return line,

def animate ( i ) :
    x = np.linspace ( 0, 12, 1000 )
    
    y = np.sin ( np.pi * ( x - 0.01 * i ) ) 
    y = y + np.sin ( 3 * np.pi * ( x - 0.01 * i ) )/3
    y = y + np.sin ( 5 * np.pi * ( x - 0.01 * i ) )/5
    y = y + np.sin ( 7 * np.pi * ( x - 0.01 * i ) )/7
    y = y + np.sin ( 9 * np.pi * ( x - 0.01 * i ) )/9
    
    line.set_data ( x, y )
    return line,

def circle ( i ) :
    # Equation of circle : x^2 + y^2 = r
    #theta = math.acos ( i/r )
    #print (i)
    x = np.linspace ( -2, 2, 1000 )
    #x = r * math.cos ( theta * math.pi/180 )
    y = math.sqrt ( r - x [ i ] * x [ i ] )#r * math.sin ( theta * math.pi/180 )
    
    #print ( theta )
    #line.set_data ( 0, 0 )
    line.set_data ( x, y )
    return line,
    

anim = FuncAnimation ( fig, circle, init_func = init, frames = 200, interval = 20, blit = True )
HTML ( anim.to_html5_video () )
#plt.close ()
#anim.save ( 'sine_wave.gif' , writer = 'imagemagick' )
```
